## Functions to calculate the roughness metric for a given CSV.

#### Helper Functions

```python
# Return 1 if t1 is first, 2 if t2 is first, 0 if equal
def is_before(t1_tup, t2_tup):
  t1_s, t1_ms = t1_tup
  t2_s, t2_ms = t2_tup
  if t1_s == t2_s:
    if t1_ms == t2_ms:
      return 0
    elif t1_ms < t2_ms:
      return 1
    elif t2_ms < t1_ms:
      return 2
  elif t1_s < t2_s:
    return 1
  elif t2_s < t1_s:
    return 2
  
# Return difference between t1 and t2 in ms
def diff_in_time(t1_tup, t2_tup):
  t1_s, t1_ms = t1_tup
  t2_s, t2_ms = t2_tup
  first = is_before(t1_tup, t2_tup)
  if first == 0:
    return 0
  elif first == 1:
    diff_s = t2_s - t1_s
    diff_ms = t2_ms - t1_ms
    if diff_ms < 0:
      diff_ms += 1000
      diff_s -= 1
  elif first == 2:
    diff_s = t1_s - t2_s
    diff_ms = t1_ms - t2_ms
    if diff_ms < 0:
      diff_ms += 1000
      diff_s -= 1
  return abs((1000. * diff_s) + diff_ms)
```

The roughness metric considers the readings in a window arround the current timepoint to gauge the overall terrain roughness of the region. Thus, need to find which readings correspond to a window around the current reading. If `window` is 50 ms, this means we are looking for the readings 50 ms before and after the current reading for a total of 100 ms.

```python
def time_window(r_metric, index, window):
  t1_tup = r_metric['time_tuples'].iloc[index]
  
  # find min_index of window
  min_index = index
  min_time = t1_tup

  while diff_in_time(t1_tup, min_time) <= window and min_index > 0:
    min_index -= 1
    min_time = r_metric['time_tuples'].iloc[min_index]

  # while loop stops once conditions are no longer met, go back one step
  if min_index > 0:
    min_index += 1
    min_time = r_metric['time_tuples'].iloc[min_index]

  # find max_index of window
  max_index = index
  max_time = t1_tup

  while diff_in_time(t1_tup, max_time) <= window and max_index < len(r_metric) - 1:
    max_index += 1
    max_time = r_metric['time_tuples'].iloc[max_index]
  
  # while loop stops once conditions are no longer met, go back one step
  if max_index < len(r_metric):
    max_index -= 1
    max_time = r_metric['time_tuples'].iloc[max_index]
  
  return min_index, max_index
```

#### Calculate the roughness metric

Calculate the discrete group of the roughness metric.
- Group 1: 0 - 2.5
- Group 2: 2.5 - 5
- Group 3: 5 - 7.5
- Group 4: 7.5+

```python
def roughness_group(x):
  if x < 2.5:
    return 0
  elif x < 5:
    return 1
  elif x < 7.5:
    return 2
  elif x >= 7.5:
    return 3
```

Given a DataFrame with z-axis acceleration, create a new column with the roughness metric for each entry.

```python
def add_roughness_metric(r_metric):
  near_vals = []
  vals = r_metric['az'].to_numpy()

  window = 500    # number of ms on either end

  for i in range(0, len(r_metric)):
    if i % 10000 == 0:
      print(i, "/", len(r_metric))

    min_index, max_index = time_window(r_metric, i, window)
    near_vals.append(vals[min_index:max_index])

  r_metric['near_vals'] = near_vals
  r_metric['near_val_std'] = [np.std(np.array(x)) for x in r_metric.near_vals.values]
  r_metric['group'] = r_metric['near_val_std'].apply(roughness_group)

  print(r_metric['near_val_std'].describe())
  print(r_metric['group'].describe())

  return r_metric
```

## Create roughness metric CSVs for each `accelerometer_calibrated_split.csv`.

```python
def plot_metric(min_index, max_index, r_metric, column_name):
  # filtered = r_metric[r_metric.index <= max_index]
  # filtered = filtered[filtered.index >= min_index]
  filtered=r_metric[min_index:max_index]
  times = filtered['utc_s']
  times += (filtered['utc_ms'] / 1000.)

  az, = plt.plot(times, filtered['az'])
  met, = plt.plot(times, filtered[column_name])
  plt.legend([az, met], ['az', column_name])
  plt.title(column_name)
  plt.show()
```

```python
# get all the accelerometer and gps csvs
filepath = '/gdrive/My Drive/Labeling/SensorData/Virb'
for root, dirs, files in os.walk(filepath):
  for f in files:
    if f == 'accelerometer_calibrated_split.csv':

      print("Creating roughness metric CSV for", os.path.split(root)[1])

      acc_csv = os.path.join(root, f)
      acc_df = pd.read_csv(acc_csv)
      r_metric = pd.DataFrame()
      r_metric['utc_s'] = acc_df['utc_s (s)']
      r_metric['utc_ms'] = acc_df['utc_ms (ms)']
      r_metric['time_tuples'] = list(zip(r_metric['utc_s'], r_metric['utc_ms']))
      r_metric['az'] = acc_df['calibrated_accel_z (m/s^2)']

      r_metric = add_roughness_metric(r_metric)
      plot_metric(10200, 11200, r_metric, 'group')

      r_metric_csv = pd.DataFrame()
      r_metric_csv['utc_s'] = r_metric['utc_s']
      r_metric_csv['utc_ms'] = r_metric['utc_ms']
      r_metric_csv['z_accel (m/s^2)'] = r_metric['az']
      r_metric_csv['roughness_group'] = r_metric['group']
      r_metric_csv.to_csv(os.path.join(root, "roughness_metric.csv"))
```
